work_year:The year the salary was paid.
experience_level: The experience level in the job during the year.
   * EN > Entry-level / Junior
   * MI> Mid-level / Intermediate
   * SE > Senior-level / Expert
   * EX > Executive-level / Director

employment_type: The type of employment for the role.
   * PT > Part-time
   * FT > Full-time
   * CT > Contract
   * FL > Freelance

job_title: The role worked in during the year.
salary: The total gross salary amount paid.
salary_currency: The currency of the salary paid as an ISO 4217 currency code.
salary_in_usd: The salary in USD.
employee_residence: Employee's primary country of residence during the work year as an ISO 3166 country code.
remote_ratio:The overall amount of work done remotely.
company_location: The country of the employer's main office or contracting branch.
company_size: The median number of people that worked for the company during the year.
 * S > Small
 * M > Medium
 * L > Large

## Data preproccessing and Model

From importing and feature desciption we can see that some of them are categorical types, such as experience level which has natural order Entry-level < Mid-level < Senior-level < Executive-level and then company size Small < Medium < Large. Then employment type, company location has no order

We will also remove columns like work_year, salary_currency and other because in my opinion they does not contribute or influence salary_in_usd

Then will encode our categorical data using LabelEncoder from scikit, because our model can accept numerical values. 

Then we will divide data into 3 sets. Train is used for training our model using current hyperparameters. Validation is used for knowing how good our model. Test is used only in the end, when we have the best hyperparameters.

```python
label = 'salary_in_usd'


# Exclude 1% of smallest and 1% of highest salaries to remove outliers
P = np.percentile(df[label], [1, 99])
df = df[(df[label] > P[0]) & (df[label] < P[1])]

df = df.drop(['work_year', 'salary_currency','employee_residence','employment_type_full', 'salary'], axis=1)
```

```python
experience_level_mapping = {'EN': 1, 'MI': 2, 'SE': 3, 'EX': 4}
df['experience_level_encoded'] = df['experience_level'].map(experience_level_mapping)

company_size_mapping = {'S': 1, 'M': 2, 'L': 3}
df['company_size_encoded'] = df['company_size'].map(company_size_mapping)
```

```python
df = df.drop(['experience_level','company_size'], axis=1)
```

```python
y = df.salary_in_usd
X = df.drop(['salary_in_usd'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

X['employment_type'] = encoder.fit_transform(X['employment_type'])
X['company_location'] = encoder.fit_transform(X['company_location'])
X['job_title'] = encoder.fit_transform(X['job_title'])
```

```python
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn import metrics

random_seed = 666

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=random_seed)
Xtrain, Xval, ytrain, yval = train_test_split(Xtrain, ytrain, test_size=0.25, random_state=random_seed)
```

#### Loss function

As a measure of dispersion, we most commonly use Mean Squared Error (MSE) or Mean Absolute Error (MAE) in our case, defined as:
$$
\quad \text{MAE} = \frac{1}{n} \sum_{i=1}^n |Y_i - \bar Y|.
$$

Key points:
MAE is less sensitive to outliers compared to Mean Squared Error (MSE) because it uses the absolute differences
MAE is in the same unit as the original data, making it easy to interpret
MAE does not penalize large errors as heavily as MSE. It treats all errors with equal importance, making it suitable when all errors should contribute equally to the overall assessment of model performance

```python
from sklearn.tree import DecisionTreeRegressor

train_acc_tree = []
val_acc_tree = []

for params in param_comb:
    current_tree = DecisionTreeRegressor(**params)
    current_tree.fit(Xtrain,ytrain)
    #train_acc_tree.append(metrics.mean_squared_error(ytrain, current_tree.predict(Xtrain), squared=False))
    #val_acc_tree.append(metrics.mean_squared_error(yval, current_tree.predict(Xval), squared=False))
    train_acc_tree.append(metrics.mean_absolute_error(ytrain, current_tree.predict(Xtrain)))
    val_acc_tree.append(metrics.mean_absolute_error(yval, current_tree.predict(Xval)))
```

From the plot below we can see how our model was overfitting with increasing depth, because of train MAE rapid decrease whereas Validation MAE was not that good. 

```python
plt.figure(figsize=(20,6))
plt.plot(train_acc_tree,'or-')
plt.plot(val_acc_tree,'ob-')
plt.xlabel('params combination')
plt.ylabel('MAE')
plt.legend(['train', 'validation'])
best_params_tree = param_comb[np.argmin(val_acc_tree)]

print(f"Best params: {best_params_tree} with MAE {min(val_acc_tree)}")
```

From the tree below we can see how our tree is choosing between features. This is one of the advantages of DecisionTree, because we can easily understand the process of deciding, how to divide our data and how to predict our salary

```python
from sklearn.tree import plot_tree

best_model = DecisionTreeRegressor(**best_params_tree).fit(Xtrain, ytrain)

# convert ndarray to dataframe again
Xtrain_df = pd.DataFrame(Xtrain, columns=X.columns)

plt.figure(figsize=(15,15))
plot_tree(best_model, filled=True, feature_names=Xtrain_df.iloc[:,:].columns, class_names=True, rounded=True, max_depth=2)
```

### Decision Boundaries

I think this could not be accurate representation of best model, because best_model uses all features, but we can see how DecisionTree tries to "cut" through vector space and divide certain data. 

I chose two features, company location and experience level, because I think they are the most important ones

```python
from matplotlib.colors import ListedColormap
cmap_bold = ListedColormap(['#0000ff', '#ff0000'])
cmap_light = ListedColormap(['#AAAAff', '#ffaaAA'])

# Code taken from ML1 notebook
def plot_decision_boundary(model, title, mesh_size = 0.5, plot_predict = False):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = Xtrain[:, 3].min() - 1, Xtrain[:, 3].max() + 1
    y_min, y_max = Xtrain[:, 4].min() - 1, Xtrain[:, 4].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, mesh_size),
                         np.arange(y_min, y_max, mesh_size))
    current_model = DecisionTreeRegressor(**best_params_tree)
    current_model.fit(Xtrain[:, [3,4]], ytrain)
    Z = current_model.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(6,6))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    if plot_predict:
        plt.scatter(Xval[:, 0], Xval[:, 1], c=current_model.predict(Xval), cmap=cmap_bold)
    else:
        plt.scatter(Xtrain[:, 0], Xtrain[:, 1], c=ytrain, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel('Company location')
    plt.ylabel('Experience level')
    plt.title(title)
```
